```python
import plotly.express as px
df = px.data.gapminder()
# print(df)
px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as nd
# df = nd.array(df)
# print(df[:,0])
fig = px.scatter_3d(df,
                 x="x",
                 y="y",
                 z="z",
            animation_frame="t"
)
fig.add_trace(
    go.Scatter3d(
            x = df["x"],
            y = df["y"],
            z = df["z"],
            mode = 'lines',
            name = 'MOON',
            marker=dict(
                size=10,  # 
                opacity=0.5 # 
            )
        ) 
)

fig.add_trace(
    go.Scatter3d(
            x = [0],
            y = [0],
            z = [0],
            mode = 'markers',
            name = 'EARTH',
            marker=dict(
                size=10,  # 
                opacity=0.5 # 
            )
        ) 
)
fig.update_layout(
            scene = dict(
                    xaxis = dict(range=[-500000,500000],),
                    yaxis = dict(range=[-500000,500000],),
                    zaxis = dict(range=[-500000,500000],)
            ),
            height = 1000,
            width = 1000,
)
fig.show()
```

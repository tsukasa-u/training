```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 100
init   = [N*0.9,N*0.05,N*0.05]
t_span = [0.0,10.0]
t_eval = np.linspace(*t_span,1000) # time for sampling
β = 1.5; γ = 0.1;
def lorenz(t,SIR,N,β,γ): 
    S,I,R = SIR
    return [-β*I*S/N, +β*I*S/N-γ*I, γ*I]
sol = solve_ivp(lorenz,t_span,init,method='RK45',t_eval=t_eval,args=(N,β,γ,))
```

```python
S,I,R = sol.y
plt.plot(S)
plt.plot(I)
plt.plot(R)
plt.legend(['S','I','R'])
```

```python
init   = [1.0, 0.0, 1.0]
t_span = [0.0,30.0]
t_eval = np.linspace(*t_span,1000) # time for sampling
def f(t,X): 
    r, v_r, v_t = X
    return [v_r, v_t**2.0/r - 1.0/r**2.0, -v_r*v_t/r + 0.01]
sol = solve_ivp(f,t_span,init,method='RK45',t_eval=t_eval,args=())

S,I,R = sol.y
plt.plot(S)
plt.plot(I)
plt.plot(R)
plt.legend(['S','I','R'])
```

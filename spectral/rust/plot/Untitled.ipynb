```python
import matplotlib.pyplot as plt
import numpy as np

# 正五角形の辺の長さ
a = 1

# 一つの正五角形を描画する関数
def draw_pentagon(x, y):
    theta = np.linspace(0, 2*np.pi, 6, endpoint=False)
    px = x + a*np.cos(theta)
    py = y + a*np.sin(theta)
    plt.fill(px, py, facecolor='gray', edgecolor='black')

# 平面充填の描画
fig, ax = plt.subplots(figsize=(6,6))
for i in range(6):
    for j in range(6):
        if i % 2 == j % 2:
            draw_pentagon(i*a + (j % 2)*a/2, np.sqrt(3)*j*a/2)

plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 正五角形の辺の長さ
a = 1

# 一つの正五角形を描画する関数
def draw_pentagon(x, y):
    theta = np.linspace(0, 2*np.pi, 5, endpoint=False)
    px = x + a*np.cos(theta + np.pi/2)
    py = y + a*np.sin(theta + np.pi/2)
    plt.fill(px, py, facecolor='gray', edgecolor='black')

# 平面充填の描画
fig, ax = plt.subplots(figsize=(6,6))
for i in range(-6, 7):
    for j in range(-6, 7):
        if (i+j) % 3 == 0:
            draw_pentagon((2*j + i)*a, np.sqrt(3)*i*a)

plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
from PIL import Image, ImageDraw

# 画像のサイズと背景色
size = (600, 600)
bg_color = (255, 255, 255)

# 画像の作成
im = Image.new("RGB", size, bg_color)

# 描画オブジェクトの作成
draw = ImageDraw.Draw(im)

# 頂点の座標を定義
vertices = [(162, 0), (438, 0), (600, 225), (375, 600), (225, 600)]

# 頂点と辺を描画
for i, v in enumerate(vertices):
    draw.ellipse((v[0]-10, v[1]-10, v[0]+10, v[1]+10), fill=(255,0,0))  # 頂点を描画
    for j in range(i+1, len(vertices)):
        draw.line((v[0], v[1], vertices[j][0], vertices[j][1]), fill=(0,0,0), width=2)  # 辺を描画

# 多角形を塗りつぶすために、頂点を時計回りに並べ替える
vertices = vertices[1:] + vertices[:1]

# 五角形を塗りつぶす
draw.polygon(vertices, fill=(255, 192, 0))

# 画像を表示
im.show()

```

```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('file.txt')
```

```python
idx = df.idx
t = df.t
r = df.x1
vr = df.x2
vt = df.x3
ur1 = df.u1
ur2 = df.u2
ut1 = df.u3
ut2 = df.u4
```

```python
ur = ur1 - ur2
ut = ut1 - ut2
df['ur'] = ur
df['ut'] = ut
```

```python
# df.plot(x='t', y=['x1', 'x2', 'x3'], subplots=True)
df.plot(x='t', y=['x1', 'x2', 'x3'], grid=True, label=['r', '$v_r$', '$v_\\theta$'])
```

```python
df.plot(x='t', y=['ur', 'ut'], grid=True, label=['$u_r$', '$u_\\theta$'])
```

```python
theta = integrate.cumtrapz(vt / r, t, initial=0)
```

```python
df['x'] = r*np.cos(theta)
df['y'] = r*np.sin(theta)
```

```python
figure, axes = plt.subplots()

axes.add_artist(plt.Circle( (0.0, 0.0), 4.0, ls="--", fill = False))
axes.add_artist(plt.Circle( (0.0, 0.0), 1.0, ls="--", fill = False))

plt.plot(df['x'], df['y'])
plt.title( 'Trajectory' )

plt.axis('equal')
axes.set_xlim(-4, 4)
axes.set_ylim(-5, 5)
plt.grid()
plt.show()
```
